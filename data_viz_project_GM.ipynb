# Data Analysis Using Vizualition Techniques

# Main motive of this project is how vizualization helps in undarstanding the data to analyse it.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# FirstDataset

```python
cr=pd.read_csv('C:/Users/cmscomputers/Downloads/Skype/Datasets/Credit (1).csv')
```

```python
cr=cr.drop(['Unnamed: 0','ID'],axis=1)
```

```python
cr
```

# Dividing features in categorical and Numeric

```python
cat=[]
quan=[]
for i in cr.columns:
    if(cr[i].dtype=='object'):
        cat.append(i)
    else :
        quan.append(i)
```

# Univariate

```python
plt.figure(figsize=(30,20))
plt.plot(cr.Income,c='purple')
plt.xlabel('Income',size=50,c='blue')
plt.xticks(range(0,420,20),size=20,c='red')
plt.yticks(size=20,c='green')
plt.show()
```

# Bivariate Analysis

```python
for i in cr.columns:
    if(cr[i].dtypes=='object'):
        sns.countplot(cr[i],hue=cr['Ethnicity'])
        plt.show()
    else:
        sns.boxplot(cr['Ethnicity'],cr[i])
        plt.show()
```

```python
sns.pairplot(cr)
```
