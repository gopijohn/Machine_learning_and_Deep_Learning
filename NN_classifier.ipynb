```python
import pandas as pd
```

```python
A=pd.read_csv('C:/Users/cmscomputers/Downloads/Skype/Loan_project/training_set (1).csv')
```

```python
A
```

```python
A.drop(['Loan_ID'],axis=1,inplace=True)
```

```python
cat=[]
con=[]
for i in A:
    if(A[i].dtype=='object'):
        cat.append(i)
    else:
        con.append(i)       
```

```python
A.isna().sum()
```

```python
for i in cat:
    c=A[i].mode()[0]
    A[i]=A[i].fillna(c)
    
for i in con:
    c=A[i].mean()
    A[i]=A[i].fillna(c)
```

```python
A.isna().sum()
```

```python
cat.remove('Loan_Status')
```

```python
Acat=pd.get_dummies(A[cat])
```

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
Acon=pd.DataFrame(mm.fit_transform(A[con]),columns=A[con].columns)
```

```python
Atr=pd.concat([Acon,Acat],axis=1)
```

```python
Atr
```

```python
X=Atr
```

```python
X.shape
```

```python
Y=A['Loan_Status']
```

```python
from sklearn.preprocessing import LabelEncoder
la=LabelEncoder()
Y_scale=pd.DataFrame(la.fit_transform(Y),columns=['Loan_Status'])
```

```python
Y_scale
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y_scale,test_size=0.2,random_state=31)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
nn=Sequential()
nn.add(Dense(614,activation='relu',input_dim=(20)))   #----input Layer
nn.add(Dense(614))                     #-----1 Hidden Layer with 614 nodes
nn.add(Dropout(0.2))

nn.add(Dense(614))                     #-----2 Hidden Layer with 614 nodes
nn.add(Dropout(0.2))

nn.add(Dense(614))                     #-----3 Hidden Layer with 614 nodes
nn.add(Dropout(0.2))

nn.add(Dense(1,activation='sigmoid'))                #----output Layer 1 node 
```

```python
nn.compile(loss='binary_crossentropy',metrics='accuracy',optimizer='adam')
nn.fit(xtrain,ytrain,epochs=500,validation_split=0.2)
```

```python
Q1 = []
for i in nn.predict(xtrain):
    if(i<0.5):
        Q1.append(0)
    else:
        Q1.append(1)
```

```python
Q2 = []
for i in nn.predict(xtest):
    if(i<0.5):
        Q2.append(0)
    else:
        Q2.append(1)
```

```python
from sklearn.metrics import accuracy_score
tr=accuracy_score(ytrain,Q1)
```

```python
tr
```

```python
ts=accuracy_score(ytest,Q2)
```
